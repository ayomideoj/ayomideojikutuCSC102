## PROJECT 1
## Pseudocode for Project 1

### INPUT number
### LET   number = n
### PRINT n-17
### IF    n>17
### PRINT n*2
### END

```python
n=int(input("Enter number"))
if n>17:
    print(2*abs((17-(n))))
    
else:
    print((n)-17)
```
